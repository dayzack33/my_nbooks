```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head(5)
```

```python
avocado.info()
```

```python
avocado['Date'].apply(pd.to_datetime)
```

```python
avocado.info()
```

```python
avocado.query(('type == "conventional"')and('year == 2015')and('region == "Albany"'))[['Date','Total Bags','Small Bags']].head(5)
```

```python
avocado_melter = avocado.melt(
    id_vars = 'Date',
    value_vars = ['Total Bags','Small Bags'],
    var_name = 'Bags',
    value_name = 'Count')
    
```

```python
avocado_melter
```

```python
sns.relplot(data=avocado_melter,kind = 'line', x='Date',y='Count',hue = 'Bags')
```

```python
pd.qcut(avocado['Total Volume'], q = 4,
        labels=['Poor','Modest','Good','Excellent'])
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q = 4,
        labels=['Poor','Modest','Good','Excellent'])
```

```python
avocado.head(5)
```

```python
sns.countplot(x=avocado.year,data=avocado,hue=avocado['Sales Volume'])
```

```python
avocado_melter2 = avocado.melt(
    id_vars = 'year',
    value_vars = ['Small Bags','Large Bags', 'XLarge Bags'],
    var_name = 'Bags',
    value_name = 'Count')
```

```python
avocado_melter2
```

```python
sns.relplot(data=avocado_melter2, kind = 'line', x='year', y='Count', hue = 'Bags')
```

**The Small Bags sell the best according to the chart above**
